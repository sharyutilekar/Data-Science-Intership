#Box Plot

```python
marks = [67,63,70,75,72,56,51,21,95,200]
```

```python
plt.boxplot(marks);
```

```python
q3 = np.percentile(marks,75)
q1 = np.percentile(marks,25)
```

```python
q3,q1
```

```python
iqr = q3 - q1
iqr #interquartile range
```

```python
#lower Bound
lb = q1 - 1.5 * iqr

#Upper Bound
up = q3 + 1.5 * iqr
```

```python
lb,up
```
